```python
import requests
import json
import pandas as pd
import time
```

```python
response900 = []

##Try to get 900 rows
for i in range(0,2000):
    offset = str((i*300)+1) #1, 301, 601
    #print(f"https://api.opensea.io/api/v1/collections?offset={offset}&limit={limit}")
    url_f = (f"https://api.opensea.io/api/v1/collections?offset={offset}&limit=300")
    response = requests.request("GET", url_f)
    response900.append(response)
    print(offset)
    time.sleep(0.001) #5 seconds of sleep
```

```python
#print(len(response900))
info_900 = []
#rinfo = json.loads(response.text)
for i in range(0,3):
    info_900.append(json.loads(response900[i].text))
#print(info_900)
```

```python
#for 900 rows
#pd.concat([df1, df2])

#df = pd.json_normalize(rinfo, record_path = ['collections'])

for i in range(0,len(info_900)):
    tempdf = pd.json_normalize(info_900[i], record_path = ['collections'])
    if i == 0:
        df_900 = tempdf
    else:
        df_900 = pd.concat([df_900, tempdf])
#df_900 = pd.json_normalize(info_900[0], record_path = ['collections'])
display(df_900)
```
